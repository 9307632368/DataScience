# series

```python
l=[11,22,33,44,55,66]
s=pd.Series(l)
print(s)
print()
x=pd.Series(l,index=["a","b","c","d","e","f"])
print(x)
```

```python
y=["rohan",69,55,7.5,"pol"]
series=pd.Series(y)
print(series)
```

```python
z=[1,2,3,4,5]
series_1=pd.Series(z,index=[1,2,3,4,5],dtype="float",name="python")
print(series_1)
```

```python


dict={"name":["java","python","c","c++"],"duration":[3,3,1,2],"rank":[3,2,1,2]}
series=pd.Series(dict)
print(series)
print(type(series))
```

```python
# addition
s1=pd.Series([11,22,33,44,55,66,77,88,99])
s2=pd.Series([2,3,4,4,3])
print(s1+s2)
```

# DataFrame

```python
l=[1,2,3,4,5,6,7,8,9]
df=pd.DataFrame(l,dtype="float")
print(df)
print(type(df))
```

```python
dict={"name":["java","python","c","c++"],"duration":[2,3,1,2],"rate":[3,3,2,1]}
df=pd.DataFrame(dict)
print(df)
print()
```

```python


dict={"name":["java","python","c","c++"],"duration":[12,11,10,9],"rating":[2,3,2,3]}
df=pd.DataFrame(dict,columns=["name"])
print(df)
print()
print(df["name"][3])
```

```python
list1=[[11,22,33,44,55],[1,2,3,4,5]]
df=pd.DataFrame(list1)
print(df)
print(type(df))
```

```python
series={"a":pd.Series([1,2,3,4]),"b":pd.Series([33,44,55,66,77])}
df=pd.DataFrame(series)
print(df)
```

```python
import pandas as pd
dict2={"name":["Rohan","bablu","husen","mahesh"],"age":[21,23,24,23],"job":["no","yes","yes","yes"]}
df=pd.DataFrame(dict2)
print(df)
print()
particular_columns=pd.DataFrame(dict2,columns=["name","age"])
print(particular_columns)
```

# Arithmetic operations

```python
import pandas as pd
var=pd.DataFrame({"a":[1,2,3,4,5],"b":[5,6,8,4,2]})
var["add"]=var["a"]+var["b"]
var["sub"]=var["a"]-var["b"]
var["mul"]=var["a"]*var["b"]
var["div"]=var["a"]/var["b"]
var
```
