# Handling Time-Series Data Issues

```python
import pandas as pd
import numpy as np

# Creating a dataset with missing timestamps and irregular frequency
data = {
    'Timestamp': ['2025-03-01 10:00:00', '2025-03-01 10:15:00', '2025-03-01 10:45:00',  # Missing 10:30
                  '2025-03-01 11:00:00', '2025-03-01 11:30:00'],  # Missing 11:15
    'Power_Consumption_kW': [50, 52, np.nan, 60, 65]  # Missing value at 10:45
}

df = pd.DataFrame(data)

# Convert timestamp to datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Set timestamp as index
df.set_index('Timestamp', inplace=True)

df
```
