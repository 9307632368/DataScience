```python
import pandas as pd

data = {
    'Customer_ID': [101, 102, 103, 104, 105],
    'City': ['NY', 'New York', 'new york', 'Los Angeles', 'LA'],  # Inconsistent city names
    'Subscription_Type': ['Basic', 'Premium', 'basic', 'VIP', 'vip'],  # Different case
    'Payment_Method': ['Credit Card', 'Paypal', 'credit card', 'paypal', 'Cash'],  # Different formats
    'Gender': ['Male', 'Female', 'M', 'F', 'Other']  # Different labels for the same category
}

df = pd.DataFrame(data)
df
```

# Merging similar categories

a) standardizing city names

```python
df['City']=df['City'].str.lower().replace({'new York':'new york',
                                  'new york': 'new york',
                                  'ny':'new york',
                                  'los angeles': 'los angeles',
                                  'la':'los angeles'
                                 })
df
```

```python
df['Subscription_Type']=df['Subscription_Type'].str.lower().replace({'basic':'basic',
                                                                    'premium':'premium',
                                                                    'vip':'vip'})
df
```

```python
df['Payment_Method']=df['Payment_Method'].str.lower().replace({
    'credit card':'credit card',
    'paypal': 'paypal',
    'cash':'cash'
})
df
```

```python
df['Gender']=df['Gender'].str.lower().replace({
    'm':'male',
    'f':'female'
    
})
df
```

# 2. Encoding Categorical Data


```python
mapping_subscription_type={'basic':1, 'premium':2, 'vip':3}
df['Subscription_Type']=df['Subscription_Type'].map(mapping_subscription_type)
df[['Subscription_Type']]
```

```python
df=pd.get_dummies(df,columns=['City','Payment_Method'])
df
```

```python
df = pd.get_dummies(df, columns=['City', 'Payment_Method'], drop_first=True)
print("\nOne-Hot Encoded Dataset:\n", df)
```
