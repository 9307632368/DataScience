```python
import pandas as pd
import numpy as np

data = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Birth_Date': ['2000-05-10', '2025-01-15', '1980-07-20', '1995-12-05', '1890-06-30'],  # 2025 is invalid, 1890 is too old
    'Salary': [50000, -1000, 70000, 80000, 3000000],  # Negative salary, extremely high value
    'Department': ['HR', 'IT', 'Finance', 'Unknown', 'Marketing'],  # 'Unknown' is an invalid department
    'Country': ['USA', 'India', 'Germany', 'UAE', 'Mars']  # 'Mars' is an invalid country
}

df = pd.DataFrame(data)

df
```

# Methods to handle incorrect or invalid data

# 1) cross validatation ( checking logical consistency) 

a) verify birth dates

```python
from datetime import datetime
df['Birth_Date']=pd.to_datetime(df['Birth_Date'],errors='coerce')
min_date=datetime(1920,1,1)
max_date=datetime.today()

df.iloc[(df['Birth_Date']<min_date)|(df['Birth_Date']>max_date)]=np.nan
df
```

b) domain specific rules

```python
# define salary limits
df.iloc[(df['Salary']<6000)|(df['Salary']>75000)]=np.nan
df
```

# reference data check

```python
valid_departments = ['HR', 'IT', 'Finance', 'Marketing', 'Operations', 'Sales']
df.loc[~df['Department'].isin(valid_departments)]=np.nan
df
```

# Handling missing or incorrect data

```python
data = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Birth_Date': ['2000-05-10', '2025-01-15', '1980-07-20', '1995-12-05', '1890-06-30'],  # 2025 is invalid, 1890 is too old
    'Salary': [50000, -1000, 70000, 80000, 3000000],  # Negative salary, extremely high value
    'Department': ['HR', 'IT', 'Finance', 'Unknown', 'Marketing'],  # 'Unknown' is an invalid department
    'Country': ['USA', 'India', 'Germany', 'UAE', 'Mars']  # 'Mars' is an invalid country
}

df = pd.DataFrame(data)

df
```

```python
df['Birth_Date'].fillna('unknown',inplace=True) # filling missing value
```

```python
# replace invalid salary with median 
df['Salary'].fillna((df['Salary']<0).median(),inplace=True)
```

```python
df
```

```python
# assign default departement
df['Department'].fillna('General',inplace=True)
df
```
