# 7. Fixing Structural Errors


```python
import pandas as pd
import numpy as np

data = {
    'Full_Name': ['Alice Smith', 'Bob_Jones', 'Charlie-Brown', 'David  White', 'Eve_King'],  # Different name formats
    'Date_of_Joining': ['2022/05/10', '10-05-2022', '2022.05.10', 'May 10, 2022', '2022-05-10'],  # Different date formats
    'Salary(USD)': ['$50,000', '60000$', '70000', '80,000', '90000'],  # Inconsistent salary formats
    'Department_Location': ['HR - NY', 'IT - SF', 'Finance - LA', 'HR-NY', 'IT, SF']  # Misaligned department-location info
}

df = pd.DataFrame(data)

df
```

# 1. Splitting and Merging Columns

```python
df[['Department','Location']]=df['Department_Location'].str.split('-',expand=True)
df.drop(columns=['Department_Location'],inplace=True)
df
```

```python
# df[['Name','Last Name']]=df['Full_Name'].str.replace('-','').str.replace('_','').str.split(' ',n=1,expand=True)
df.drop(columns=['Full_Name'],inplace=True)
df
```

# 2. Parsing Incorrect Data Entries

```python
df['Salary(USD)'].unique()
```

```python
df['Salary(USD)']=df['Salary(USD)'].replace({'\$':'',',':''},regex=True).astype(float)
df
```
