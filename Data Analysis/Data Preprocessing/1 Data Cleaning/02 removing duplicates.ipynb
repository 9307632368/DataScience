# removing duplicates

```python
import pandas as pd
```

```python
data=pd.read_csv('datasets/student_admission_record.csv')
```

```python
data.info()
```

```python
data.describe()
```

# detecting duplicates

```python
# check for duplicates
duplicated=data.duplicated().sum()
duplicated
```

```python
# show all duplicated rows
data[data.duplicated()]
```

```python
# find the duplicates based on specific columns
data[data.duplicated(subset=['Name','Gender','High School Percentage'])]
```

# removing duplicates

```python
# remove all duplicates
data.drop_duplicates(inplace=True)
```

```python
# remove duplicates based on specific columns
data.drop_duplicates(subset=['Name','Gender','High School Percentage'],inplace=True)
```

```python
# remove duplicates and keep last occurance
data.drop_duplicates(keep='last')
```
