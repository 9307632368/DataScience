# 8. Data Deduplication and Normalization

```python
import pandas as pd
import numpy as np

# Creating a dataset with structural errors
data = {
    'Full_Name': ['Alice Smith', 'Bob_Jones', 'Charlie-Brown', 'David  White', 'Eve_King'],  # Different name formats
    'Date_of_Joining': ['2022/05/10', '10-05-2022', '2022.05.10', 'May 10, 2022', '2022-05-10'],  # Different date formats
    'Salary(USD)': ['$50,000', '60000$', '70000', '80,000', '90000'],  # Inconsistent salary formats
    'Department_Location': ['HR - NY', 'IT - SF', 'Finance - LA', 'HR-NY', 'IT, SF']  # Misaligned department-location info
}

df = pd.DataFrame(data)

df
```

```python
df.duplicated().sum()
```

```python
# df.drop_duplicates(inplace=True)
df
```

```python
df['Salary(USD)']=df['Salary(USD)'].replace({'\$':'',',':''},regex=True).astype(float)
```

# 2. Normalization (Scaling Numerical Data)

```python
df['Salary(USD)_Normalized']=(df['Salary(USD)']-df['Salary(USD)'].min())/(df['Salary(USD)'].max()-df['Salary(USD)'].min())
```
