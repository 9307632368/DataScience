### (A) Train-Test Split


```python
import pandas as pd
from sklearn.model_selection import train_test_split
data=pd.read_csv('C:/Users/polro/06 data preprocessing/datasets/titanic.csv')
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

### (B) Train-Validation-Test Split

```python
# split data into 60% for training ,20% validation and 20% for testing
x_train,x_temp,y_train,y_temp=train_test_split(x,y,test_size=0.4,random_state=42)
x_val,x_test,y_val,y_test=train_test_split(x_temp,y_temp,test_size=0.5,random_state=42)
print(x_train)
print(x_val)
print(x_test)
```

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=5,shuffle=True,random_state=42)
for train_index,test_index in kf.split(x):
    x_train,x_test=x.iloc[train_index],x.iloc[test_index]
    y_train,y_test=y.iloc[train_index],y.iloc[test_index]
    print(x_train)
```

### (D) Stratified Split (For Imbalanced Data)

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=42,stratify=y,random_state=42)
print(x_train.value_counts())
print()
print(y_train.value_counts())
```
