# 1. Normalization (Scaling Data to a Range)

```python
import pandas as pd
import numpy as np

# Creating a sample dataset
data = {'Feature': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)
df
```

```python
df['feature_normalized']=(df['Feature']-df['Feature'].min())/(df['Feature'].max()-df['Feature'].min())
df
```

# 2. Standardization 

```python
df['Feature_standarized']=(df['Feature']-df['Feature'].mean())/df['Feature'].std()
df
```

# 3. Log Transformation (Handling Skewed Data)

```python
df['Feature_log']=np.log(df['Feature'])
df
```

# 4. power transformation

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson')
df['Feature_Power'] = pt.fit_transform(df[['Feature']])
```

# 5. Binning (Converting Continuous Data to Categorical)

```python
df['Feature_Binned'] = pd.cut(df['Feature'], bins=3, labels=['Low', 'Medium', 'High'])
df
```
