# Feature Engineering 

### 1. Types of Feature Engineering

#### (A) Handling Missing Values

```python
import pandas as pd
import numpy as np

data = {'Age': [25, np.nan, 30, 35, np.nan, 40]}
df = pd.DataFrame(data)
df['Age'].fillna(df['Age'].mean(),inplace=True)
df
```

#### (C) Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
data = {'Salary': [20000, 50000, 80000, 100000]}
df = pd.DataFrame(data)
mms=MinMaxScaler()
df['Salary_Normalized']=mms.fit_transform(df[['Salary']])
df

# another method without using package
df['Salary_Normalized_2']=(df['Salary']-df['Salary'].min())/(df['Salary'].max()-df['Salary'].min())
df
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
df['Salary_Standardization']=ss.fit_transform(df[['Salary']])

# another method without using an
df['Salary_Standardization_2']=(df['Salary']-df['Salary'].mean())/df['Salary'].std()
df
```

#### (D) Feature Creation (Feature Extraction)

```python
df = pd.DataFrame({'Date': pd.to_datetime(['2023-01-15', '2023-06-10', '2023-09-25'])})
df
```

```python
df['Year']=df['Date'].dt.year
df['Month']=df['Date'].dt.month
df['Day']=df['Date'].dt.day
df['Week']=df['Date'].dt.dayofweek
df
```

#### (E) Feature Transformation


```python
df = pd.DataFrame({'Income': [1000, 5000, 10000, 50000, 100000]})
#log transformation
df['Income_Log_Transformation']=np.log(df['Income'])
df
```

# 2. Feature Selection (Choosing the Best Features)

#### (A) Removing Low Variance Features

```python
from sklearn.feature_selection import VarianceThreshold
df = pd.DataFrame({'Feature1': [1, 1, 1, 1], 'Feature2': [1, 2, 3, 4]})
vt=VarianceThreshold(threshold=0.1)
df.iloc[:,vt.fit(df).get_support()]
```

#### (B) Correlation-Based Feature Selection
