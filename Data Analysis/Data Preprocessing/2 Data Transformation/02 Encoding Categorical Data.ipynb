# Encoding Categorical Data

# 1. Label Encoding

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = {'Weather': ['Sunny', 'Cloudy', 'Rainy', 'Sunny', 'Rainy']}
df = pd.DataFrame(data)
df
```

```python
le=LabelEncoder()
df['Wheather_Encoded']=le.fit_transform(df['Weather'])
df
```

# 2. One Hot Encoder

```python
data = {'Color': ['Red', 'Blue', 'Green', 'Blue', 'Red']}
df = pd.DataFrame(data)
df[['Color_Red','Color_Blue','Color_Green']]=pd.get_dummies(df)
df.drop(columns='Color',inplace=True)
df
```

```python
from sklearn.preprocessing import OneHotEncoder
data = {'Color': ['Red', 'Blue', 'Green', 'Blue', 'Red']}
df = pd.DataFrame(data)
ohe=OneHotEncoder(sparse=False)
encoded_array=ohe.fit_transform(df[['Color']])
one_hot_encode=pd.DataFrame(encoded_array,columns=ohe.get_feature_names_out(['Color']))
one_hot_encode
```
